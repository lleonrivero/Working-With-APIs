# FBI API

```python
fbi = requests.get("https://api.fbi.gov/wanted/v1/list")
print("fbi:", fbi.status_code)
```

```python
fbi_response = requests.get('https://api.fbi.gov/wanted/v1/list')
fbi_response
fbi_results = fbi_response.json()
#fbi_results
```

```python
pd.set_option("display.max_rows", 0)

fbi_data = pd.DataFrame(fbi_results)
fbi_data.head()
```

```python
def flatten(fbi_data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(fbi_data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        fbi_data = pd.concat([fbi_data, flattened], axis=1)
        fbi_data = fbi_data.drop(column, axis=1)
    return fbi_data
```

```python
dict(fbi_data['items'])
pd.DataFrame(dict(fbi_data['items']))
pd.DataFrame(dict(fbi_data['items'])).transpose().head(3)
```

```python
fbi_nested_columns = ['items']

fbi_flat = flatten(fbi_data, fbi_nested_columns)
fbi_flat.head(3)
```
